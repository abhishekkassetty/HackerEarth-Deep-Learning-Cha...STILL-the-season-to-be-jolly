```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import cv2
import PIL
import pathlib
import shutil
import sys
from keras.models import Sequential,Model
from keras_preprocessing.image import ImageDataGenerator
from keras.layers import Dense,Conv2D,Flatten,Dropout,Activation,BatchNormalization,MaxPooling2D,GlobalAveragePooling2D
from keras import regularizers,optimizers
from keras.losses import CategoricalCrossentropy
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from keras.callbacks import ModelCheckpoint
```

```
train_path = r"/content/dataset/train"
test_path = r"/content/dataset/test"
```

```
df = pd.read_csv("/content/dataset/train.csv")
```

```
train0 = r"/content/dataset/train0"
```

```
if not os.path.exists(train0):
  os.mkdir(train0)

for image_name,class_name in df.values:
  if not os.path.exists(train0 + "/" + str(class_name)):
    os.mkdir(train0 + "/" + str(class_name))
  src_path = train_path + "/" + image_name 
  des_path = train0 + "/"+ str(class_name) + "/" + image_name 
  try:
    shutil.move(src_path,des_path)
  except:
    print("error")

print("done")
```

```
categories = ["Airplane","Candle","Christmas_Tree","Jacket","Miscellaneous","Snowman"]
```

```
labels = [i for i in range(0,7)]
```

```
label_dict = {categories[i]:labels[i] for i in range(len(categories))}
```

```
data = []
target = []
```

```
for category in categories:
    filepath = os.path.join(train0,category)
    img_names=os.listdir(filepath)
    for img_name in img_names:
        img_path=os.path.join(filepath,img_name)
        img=cv2.imread(img_path)
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        resized=cv2.resize(gray,(100,100))
        data.append(resized)
        target.append(label_dict[category])
        
```

```
data = np.array(data)
data = data/255
data = np.reshape(data,(data.shape[0],100,100,1))
```

```
target = np.array(target)
target = to_categorical(target)
```

```
model = Sequential()
```

```
model.add(Conv2D(200,(3,3),input_shape = data.shape[1:],activation = "relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Conv2D(100,(3,3),activation = "relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Flatten())
model.add(Dense(150,activation = "relu"))
model.add(Dense(75,activation = "relu"))
model.add(Dense(6,activation = "softmax"))
```

```
model.compile(loss = "categorical_crossentropy",optimizer = "adam",metrics = ["accuracy"])
```

```
checkpoint = ModelCheckpoint("model-{epoch:03d}.model",save_best_only=True,mode="auto")
```

```
fitting = model.fit(data,target,validation_split=0.2,epochs = 10,callbacks=[checkpoint])
```

```
model.save('Image_classification01')
```
